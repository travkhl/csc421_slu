## Think/Acting  Humanly/Rationally 


1. **Think like humans**

   This approach to AI aims to better understand how humans think/accomplish various tasks. It is informed by Cognitive science which explores how humans perform various tasks. As an example of this approach we might be interested in developing computer models of how humans play chess or some other board game. The goals of these models is to better understand <mark>what strategies humans use to make decisions</mark> in this context including thier **limitations** rather than developing the "best" computer chess playing program that could use a completely different approach to playing.
<br>
  
  
2. **Think rationally**

   Aristotle established logic as a way to formalize "right thinking" (irrefutable reasoning processes) Logic deals with **relationships** and abstracts the **process of reasoning** so that it is independent of the specific categories used. For example one can have the following thinking "patterns":

   * All dogs are animals. Robin is a dog. Therefore robin is an animal.
   * Every student has a computer. Each computer has a mouse. Therefore, every students has a mouse.
   
    Such patterns can be used to <mark>derive information that we don't know from facts that we know</mark>. For example if one knows that the parents of Bart are Homer and Marge, one can infer that Marge has a child named Bart. 
<br>
  
   
3. **Act like humans**

    The first proposal for evaluating success in AI was building a program that acts humanly named after Allan Turing. To be considered intelligent a program must be able to <mark>act sufficiently like a human to fool an interrogator</mark>. A human interrogates the program and another human via a terminal simultaneously. If after a reasonable period, **the interrogator cannot tell which is which**, the program passes. Passing the Turing test requires <u>_natural language procesing, knowledge representation, automated reasoning, and machine learning_</u> all of which have been active areas of research since the dawn of AI in the 1950s. One could also add _computer vision and audition_ as well as _robotics_ as additional challenges while keeping the successful immitation aspects of the test the same.
<br> 
  

4. **Act rationally** 

    An **agent** is something that acts i.e does things. Agents operate autonomously, perceive their environment, are able to affect it, can change, and create and pursue goals. A **rational agent** is an agent that tries to <mark>achieve the best outcome or the best expected outcome when there is uncertainty</mark>. 
    
    What is **rational** is actually a relatively complicated topic that is probably best answered with the non-answer "I don't know how to explain what it is but I sure know when I see it". Examples:  
    * In some cases a very simple behavior such as closing your eyelid when you detect an object moving towards it does not require any inference and **a simple reflex** will do.  
    * In other case, more **complicated behaviors based on inference** (deriving new bits of information from existing bits of information) **and learning** (improving how tasks are performed based on exposure to novel situations).  

    However, perfect rationality is not feasible in complex environments. The computational demands are just too high. Later, in "Adverserial Search and Games", we will deal with the issue of limited rationality—**acting appropriately** when there is not enough time to do all the computations one might like.

**Humanly:** Not focus on being optimal. Instead, we may want AI to join our human society. It aims for human-like behavior or cognition, making AI agents suitable for interacting with humans naturally in daily tasks.  
**Rationally:** Focus on optimal and goal-directed behavior. We may want AI to operate independently and make AI as a tool for effectiveness—much like how airplanes achieve flight without needing to resemble birds.

### Beneficial Machines

The standard model works well only when given a fully specified objective. But in real world, like self-driving car, we can’t write down a perfect objective. For example, “be safe” conflicts with “reach the destination”, and we must balance these goals. 

So we need to <mark>align machine's objectives with human true preferences</mark>. The problem is: humans have many preferences, and we cannot list all of them. If we cannot transfer all the human objectives perfectly to the machine, then we need to introduce some form of “uncertainty” into its objective, to prevent the AI from pursuing its goal by any means necessary. <mark>If the machine knows it might misunderstand the goal, it will behave more safely</mark>: it will be cautious, ask before acting, try to learn more about our preferences, and let humans stay in control. Ultimately, we want agents that are provably beneficial to humans.

# History of AI

- **1945-1956: The inception of AI.** First artificial neurons, early learning rules, and machine reasoning systems. Dartmouth workshop that officially launched AI as a discipline. 
- **1952-1969: Early enthusiasm, great expectations.** Symbolic AI, reinforcement learning, neural networks, early theorem provers, problem solvers, and “microworld” systems.
- **1966-1973: A dose of reality.** 
    - Early AI relied heavily on “informed introspection”, rather than analyzing and planning what tools or algorithms to use. 
    - Underestimated the intractability of many problems, which were far from simple “try-and-see” combinations but instead suffered from combinatorial explosion.  

        > * _"Before the theory of computational complexity was developed, it was widely thought that “scaling up” to larger problems was simply a matter of faster hardware and larger memories."_  
        > * _Can find a solution **in principle** $\neq$ Contain any of the mechanisms needed to find it **in practice**"_  

    - Single-layer perceptrons were proven to represent only very limited functions, which led the research community and funding agencies to lose confidence in perceptron-based approaches. Although the ideas of multilayer perceptrons (MLPs) and backpropagation (BP) had already appeared in the 1960s, the lack of effective training methods, limited computational resources, and the negative perception of perceptrons caused investments and interest in neural networks to decline. This indirectly slowed down progress in AI research.

- **1969-1986: Expert systems.**
    - Weak methods: general-purpose, search-based, lacking domain knowledge. Effective in tiny toy worlds, but not scalable to complicated real-world problems.  

        > _“I don’t really know how to drive, but I know the basics—accelerate, brake, turn the wheel—so I’ll just try and see if I can get the car to the destination.”_   

    - Knowledge intensive expert systems: rely on domain-specific rules to make large reasoning steps.
        
        > _“I understand traffic rules, road conditions, common driving scenarios, and practical experience, so I know which actions are most likely to succeed.”_   
        
        - not generalizable
        - cannot understand natural language
        - knowledge bases are difficult to build and expensive to maintain
        - cannot deal with uncertainty
        - do not learn from experience / lack machine learning (ML)
- **1986-Present: The return of neural networks.** In the mid-1980s at least 4 different groups reinvented BP learning algorithm first developed in the early 1960s.  

    > * _Terrence Deacon: At some level humans manipulate symbols–this is the defining characteristic of humans._     
    > * _Geoffrey Hinton: Symbols are the “luminiferous aether of AI,” referring to a medium 19th-century physicists believed existed but was later proven nonexistent._     
    
    Many concepts expressed in natural language do not have clean, axiomatic definitions in the real world—an area where neural networks, with their ability to form flexible and internal representations, are often a better fit.
- **1987-Present: Probabilistic reasoning and machine learning.** 
- **2001-Present: Big data.**
- **2011-Present: Deep learning.**
